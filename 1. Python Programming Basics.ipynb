## Sets 
Collection of elements with NO DUPLICATE values

```python
#### Creating Sets ####

# Defining sets
set1={2,2,3,1,1,4,4}
set2={5,5,3,4,5,5}
set3={'b','b','a','c','c',1,2,3}

print(type(set1))

# printing the sets
print(set1)
print(set2)
print(set3)
```

```python
# Removing duplicate values

# Defining a sample list and tuple
sampleList=[2,3,4,4,4,5,5,5,8,2,'b','b']
sampleTuple=(2,3,'a','a',4,4,10)

# Removing duplicates using set
print(list(set(sampleList)))
print(tuple(set(sampleTuple)))
```

```python
#### Common Set commands ####

# Defining sets
set1={2,2,3,1,1,4,4}
set2={5,5,3,4,5}
print('%set_env1:',set1)
print('set2:',set2)

# set OR operation (Union)
print('OR Operation:', set1 | set2)
      
# set AND operation (Intersection)
print('AND Operation:', set1 & set2)

# What elements are present in set1 which are not in set2
print('uncommon elements:', set1 - set2)
      
# What elements are present in set2 which are not in set1
print('uncommon elements:',set2 - set1)

# adding a SINGLE new item to set
set1.add(10)
print('After adding new element:',set1)

# adding a MULTIPLE new items to set
set1.update([1.5,20,30,40,10,10])
print('After adding multiple new elements:',set1)

# removing an element from the set
set1.remove(30)
print('After removing SINGLE element:',set1)

# removing MULTIPLE elements from the set
set1.difference_update([20,30,2,3])
print('After removing MULTIPLE elements:',set1)

# clear() function will remove all elements
print(set1.clear())
print(set1)
```

```python
#### Sets does not support indexing

# Defining a set
set1={2,3,4,5}

#sets does not support indexing
set1[2]
```

```python
# Common Dictionary commands

# Creating a simple dictionary
dict1={101:'ram', 102:'sham', 103:'ravi'}

# Extracting only the keys
print('All the keys in the dictionary:',dict1.keys())

# Extracting only the values
print('All the values in the dictionary:',dict1.values())

# Extracting only the values
print('All the key-value pairs in the dictionary')
print(dict1.items())

# Get the value for a particular key
print('Value for key 101:',dict1.get(101))

# Get the values for multiple keys
# No pre-built function for it, loop is needed
key_set=[101,102]
for k in key_set:
    print(dict1.get(k))
    
# Adding a new entry to the dictionary
# Keys can be of different datatypes
dict1[101]=[]
dict1[302]=[2,3,4]
dict1['ID_01']='test item'
print(dict1)

# Deleting an item from the dictionary based on keys
dict1.pop(302)
print('dictionary after removing value of key 302:', dict1.items())

# Deleting the last item from the dictionary
dict1.popitem()
print('dictionary after removing last key-value pair:', dict1.items())
```

### IF condition statement
if a condition is True, then do something

```python
# Do something only if the condition is true

# Creating a variable
x=10

if(x>5):
    
    print("condition is true!")
    
    # creating another variable
    y=100
    
    print(y)

    print("hello")
```

### if-else statement
if the condition is true, then do this, other wise do that

```python
# Creating a variable
x=10

# if the condition is not true then else statements will execute
if(x>50):
    # IF block
    print("condition is true!")

    # creating another variable
    y=100+x
    print(y)
    
else:
    # Else block
    print('The condition is not true!')
    y=400+x
    print(y)
```

### nested if-else (elif statement)

Evaluating multiple if-else conditions one after the other

```python
# Creating a variable
x=10

if(x>5):
    print("condition is true!")
    # creating another variable
    y=100
    print(y)
    
elif(x<10):
    print('x is less than 10!')
    
elif(x==10):
    print('x is equal to 10!')
    
else:
    print('this is the end game!')
```

### for loop for multiple iterators
Looping through multiple list of values together

```python
# Creating sample objects
sampleList=[10,20,30,40]
anotherList=[1,2,3]
sampleTuple=('a','b','c','d')

# Looping three objects at a time
for numIterator, i, charIterator in zip(anotherList, sampleList,  sampleTuple):
    print(numIterator, '--', i , "--", charIterator)
```

```python
# Creating sample objects
sampleList=[10,20,30]
sampleTuple=('a','b','c','d')
anotherList=[1,2,3]

# Nested for loop for multiple iterators
for numIterator in sampleList:
    for charIterator in sampleTuple:
        for i in anotherList:
            print(numIterator, '--', charIterator, '--', i)
```

```python
# Creating sample objects
sampleList=[10,20]
sampleTuple=('a','b','c','d')

# Nested for loop for multiple iterators
for charIterator in sampleTuple:
    for numIterator in sampleList:
        print( charIterator, '--', numIterator)
```

### while loop

```python
import time
# Creating a counter
counter=1

# Run till condition becomes false
while counter < 10:
    print('Before Update',counter)
    counter+=3
    # Waiting for 2 seconds
    time.sleep(2)
    print('After Update',counter)
```

### Continue
if a condition is met, don't execute the remaining statements, go to next iteration

```python
# Creating a counter
counter=1

# Run while condition is true
while counter < 10:
    
    # Checking a condition
    if(counter==7):
        counter += 1  
        # Exit the current iteration
        continue
        print(1+20)
        print('From IF:',counter)
        
    else:
        print('From else',counter)
        counter += 1
```

```python
# Another example of continue statement
# Creating sample objects
sampleList=[10,20,30,40]
sampleTuple=('a','b','c','d')

# Nested for loop for multiple iterators
for numIterator in sampleList:
    for charIterator in sampleTuple:
        # Exit the current iteration if 'b' is encounterd
        if((numIterator==30) & (charIterator=='c')):
            continue
        print(numIterator, '--', charIterator)
```

### Break
if a condition is met, break the loop

```python
counter=1
while counter < 10:
    # Evaluating the condition
    if(counter==5):
        print('the last counter value was',counter)
        break
    else:
        print(counter)
        counter += 1
```

```python
# Another example of break statement
# Creating sample objects
sampleList=[10,20,30,40]
sampleTuple=('a','b','c','d')

# Nested for loop for multiple iterators
for numIterator in sampleList:
    for charIterator in sampleTuple: 
        # Completely break the current loop if 'c' is encounterd
        if(charIterator=='c'):
            break
        print(numIterator, '--', charIterator)
```

### Try Except
try-except block is used to handle coding exceptions elegantly

```python
# With Try - Except block
y='x'

if(y=='x'):
    try:
        print(100/0)
        print('Result:',y+100)
        
    except Exception as exceptionMsg:
        print('>>>> Something went wrong!:', exceptionMsg, 'the input was', y)
        
else:
    print('Condition is not true')
```

```python
# Without try - except
y=0

if(y==0):
    print(100/10)
    print('Result:',100/y)
    
else:
    print('Condition is not true')
```
